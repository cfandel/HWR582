** Schematic plots of likelihood distributions to identify discriminatory data**


```python
#Import packages:
import numpy as np
import scipy as sp
from scipy import stats
from matplotlib import pyplot as plt
```

**Generate data**

```python
#DEFINE INPUTS HERE:

#Set head range:
Hmin = 0
Hmax = 100

#Set means & standard deviations for different scenarios:
stats = {}

stats['good','moc',  'mean']  = 25
stats['good','moc',  'stdev'] = 5
stats['good','other','mean']  = 75
stats['good','other','stdev'] = 5

stats['intermediate','moc',  'mean']  = 40
stats['intermediate','moc',  'stdev'] = 5
stats['intermediate','other','mean']  = 60
stats['intermediate','other','stdev'] = 5

stats['bad','moc',  'mean']  = 50
stats['bad','moc',  'stdev'] = 5
stats['bad','other','mean']  = 50
stats['bad','other','stdev'] = 5
```

```python
#Generate data:

#Generate heads:
heads = np.arange(Hmin,Hmax+1)

#Define possible combinations:
quality = ['good','intermediate','bad']
moc = ['moc','other']

#Generate likelihood distributions for each scenario:
L = {}
for i in quality:
    for j in moc:
        mean = stats[i,j,'mean']
        stdev = stats[i,j,'stdev']
        L[i,j] = sp.stats.norm.pdf(heads, loc=mean, scale=stdev)
```

**Plotting**

```python
#Plot
for i in quality:
    for j in moc:
        plt.plot(L[i,j])
```
